```python
# Step 1: Import required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load dataset from URL
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# Step 3: View first few rows
print("üîç First 5 rows of the dataset:")
print(df.head())

# Step 4: Summary statistics
print("\nüìä Summary Statistics (.describe()):")
print(df.describe())

print("\nüìà Mean of numeric columns:")
print(df.mean(numeric_only=True))

print("\nüìâ Median of numeric columns:")
print(df.median(numeric_only=True))

print("\nüìè Standard Deviation of numeric columns:")
print(df.std(numeric_only=True))

# Step 5: Visualize histograms for all numeric features
print("\nüìä Displaying histograms for numeric features:")
df.hist(bins=20, figsize=(15, 10), color='lightblue', edgecolor='black')
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()
```
